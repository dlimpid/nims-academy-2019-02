Comprehensionì„ ì‚¬ìš©í•˜ë©´ (ê°„ë‹¨í•œ) ë°˜ë³µë¬¸ê³¼ ë¶„ê¸°ë¬¸ì„ í›¨ì”¬ ì§§ê²Œ í‘œí˜„í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤. (ê·¸ë¦¬ê³  ë§ì€ ê²½ìš° ì„±ëŠ¥ë„ ë” ì¢‹ìŠµë‹ˆë‹¤.)

```python
[<any expression with an element> for <element> in <iterable> if <condition>]
```

````python
[n + 10 for n in range(10) if n % 2 == 0]
````

````python
a = list(range(-10000, 10000))
````

````python
%%timeit

b = []
for x in a:
    if x % 2 != 0:
        b.append(x)
````

````python
%%timeit

b = [x for x in a if x % 2 != 0]
````
