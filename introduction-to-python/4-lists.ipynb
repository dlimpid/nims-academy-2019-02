```python
desserts = [
    'Brownies',
    'Tiramisu',
    'Cakes',
    'Doughnuts',
    'Chocolate Chip Cookies',
    'Pudding',
    'Tarts',
]
```

```python
sorted_desserts = sorted(desserts)
print(sorted_desserts)
print(desserts)
```

```python
sorted_desserts = sorted(desserts, reverse=True)
print(sorted_desserts)
print(desserts)
```
